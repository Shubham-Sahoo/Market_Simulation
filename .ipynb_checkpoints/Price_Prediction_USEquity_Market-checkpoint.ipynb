```python
# Generate data point for each iteration

class gen_dataset:
    
    def __init__(self, dataset, attribute = 'Close'):
        self.attribute = attribute
        self.dataset = iter(list(dataset[:][attribute].values))
        
    def get_data(self):
        return next(self.dataset)
        
```

```python
# Create dynamic dataset 
memory_time = 30
ohlc_attr = 'Close'
training_data = []

data_obj = gen_dataset(x)

for i in range(memory_time):
    data_val = data_obj.get_data()
    training_data.append(data_val)
```

```python
type(training_data)
```

```python
# Value prediction model training using 30 days of data

model_predictions = []
N_test_observations = len(x) - memory_time - 1
print(N_test_observations)
test_data_real = []

for time_point in range(N_test_observations):
    data_val = data_obj.get_data()
    test_data = data_val
    model = ARIMA(training_data, order=(4,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    #print(output)
    yhat = output[0]
    model_predictions.append(yhat)
    #print(model_predictions)
    test_data_real.append(test_data)
    training_data.pop(0)
    training_data.append(data_val)
```

```python
test_set_range = len(test_data_real)
plt.plot(model_predictions, color='blue', marker='o', linestyle='dashed',label='Predicted Price')
plt.plot(test_data_real, color='red', label='Actual Price')
plt.title('Microsoft Prices Prediction')
plt.xlabel('Date')
plt.ylabel('Prices')
#plt.xticks(np.arange(0,252,1), x.Date[881:1259:50])
plt.legend()
plt.show()
```

```python
def calculate_metrics(cumret):
    '''
    calculate performance metrics from cumulative returns
    '''
    total_return = (cumret[-1] - cumret[0])/(cumret[0])
    apr = (1+total_return)**(252/len(cumret)) - 1
    rets = pd.DataFrame(cumret)
    sharpe = np.sqrt(252) * np.nanmean(rets) / (np.nanstd(rets))
    
    return total_return, apr, sharpe
```

## Pair Trading Strategy

```python
from datetime import date
import yfinance as yf
from pandas.plotting import lag_plot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import coint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import statsmodels.api as sm
```

```python
# List of Trending Equities

trend_equity = ["AMD", "NVDA", "FCEL", "BABA", "AAPL", "MSFT", "TSM", "TXN", "ADI", "AMAT"]

ticker_list = []
for i in trend_equity:
    ticker_list.append(yf.Ticker(i))

print(ticker_list)
```

```python
# Get closing price of each of the stocks for last 5 years

attribute = 'Close'
prices_5y = ticker_list[0].history(period='5y',interval='1d')[attribute]

for i in range(len(ticker_list)-1):
    prices_5y = pd.concat([prices_5y, ticker_list[i+1].history(period='5y',interval='1d')[attribute]], axis = 1)

prices_5y.columns = trend_equity
print(prices_5y)
```

```python
prices_5y
```

```python
# Calculate the returns
returns = np.log(prices_5y).diff().dropna()
```

```python
returns[1000:]
```

```python
form_start = '2018-06-07'
form_end = '2022-06-01'
trade_start = '2022-06-02'
trade_end = '2023-06-06'

prices_form = prices_5y[form_start:form_end]
prices_trade = prices_5y[trade_start:trade_end]
returns_form = returns.loc[form_start:form_end]
returns_trade = returns.loc[trade_start:trade_end]
```

```python
'''
# Calculate Kendall's tau for each pair of stocks
'''

results_corr = pd.DataFrame(columns=['tau'])

for s1 in returns_form.columns:
    for s2 in returns_form.columns:
        if (s1!=s2) and (f'{s2}-{s1}' not in results_corr.index):
            results_corr.loc[f'{s1}-{s2}'] = stats.kendalltau(returns_form[s1], returns_form[s2])[0]
```

```python
results_corr.sort_values(by='tau')
```

```python
def parse_pair(pair):
    s1 = pair[:pair.find('-')]
    s2 = pair[pair.find('-')+1:]
    return s1,s2
```

```python
selected_stocks = []
selected_pairs = []
num_pairs = 5 # number of pairs to select

for pair in results_corr.sort_values(by='tau', ascending=False).index:
    s1,s2 = parse_pair(pair)
    if (s1 not in selected_stocks) and (s2 not in selected_stocks):
        coint_pval = coint(prices_form[s1], prices_form[s2])[1] # ADF-test p-value
        if coint_pval<0.1: 
            selected_stocks.append(s1)
            selected_stocks.append(s2)
            selected_pairs.append(pair)
    
    if len(selected_pairs) == num_pairs:
        break

```

```python
delta = 1.645
rho = 0.7

algo_returns = pd.DataFrame(index=prices_trade.index, columns=selected_pairs)

for pair in selected_pairs:
    s1,s2 = parse_pair(pair)
    spread = prices_trade[s1] / prices_trade[s2]
    
    # calculate positions
    positions = pd.DataFrame(index=prices_trade.index, columns=[s1,s2])

    condition_short_high = (spread > res.params['const[0]'] + delta * res.params['sigma2[0]'])
    #print(condition_short_high)
    positions[condition_short_high] = [-1,1]

    condition_long_high = (spread < res.params['const[0]'] - delta * res.params['sigma2[0]'])
    positions[condition_long_high] = [1,-1]

    condition_short_low = (spread > res.params['const[1]'] + delta * res.params['sigma2[1]']) 
    positions[condition_short_low] = [-1,1]

    condition_long_low = (spread < res.params['const[1]'] - delta * res.params['sigma2[1]'])
    positions[condition_long_low] = [1,-1]

    positions = positions.fillna(0)
    
    # calculate returns
    simple_ret = (np.exp(returns_trade[[s1,s2]])-1) # transform log-returns to simple returns
    ret = (positions.shift() * simple_ret).sum(axis=1) * 2 # double capital (from short positions)
    algo_returns[pair] = ret
```

```python
algo_cumret = {}

for num_pairs in [1]:
    algo_cumret[f'{num_pairs}p'] = np.nancumprod(algo_returns[selected_pairs[:num_pairs]].sum(axis=1) 
                                                 / num_pairs + 1)
    
plt.figure(figsize=(18,6))
for k in algo_cumret.keys():
    plt.plot(algo_cumret[k], label=f'algo_{k}')
plt.legend()
```

```python
def calculate_metrics(cumret):
    '''
    calculate performance metrics from cumulative returns
    '''
    total_return = (cumret[-1] - cumret[0])/cumret[0]
    apr = (1+total_return)**(252/len(cumret)) - 1
    rets = pd.DataFrame(cumret).pct_change()
    sharpe = np.sqrt(252) * np.nanmean(rets) / np.nanstd(rets)
    
    return total_return, apr, sharpe
```

```python
metrics = pd.DataFrame(index=algo_cumret.keys(), columns=['Total return', 'APR', 'Sharpe'])
for k in algo_cumret.keys():
    metrics.loc[k] = calculate_metrics(algo_cumret[k])
```
